```python
import matplotlib.pyplot as plt

# Data
protocols = ['UDP', 'TCP', 'ESP-Mesh', 'zh-network (ESP-NOW)', 'boot+wifi', 'boot']
charges = [118.82, 137.74, 451.47, 16.41, 43.6 ,3.21]

# Create a bar plot
plt.figure(figsize=(10, 6))
bars = plt.bar(protocols, charges)

# Add value labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{height}',
             ha='center', va='bottom')

# Customize the plot
#plt.title('Current Consumption for Sending a Small Wireless Message with ESP32')
plt.xlabel('Funkprotokoll')
plt.ylabel('benötigte Ladung (mAs)')
plt.ylim(0, max(charges) * 1.2)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Save the plot
plt.savefig('plot.png', format='png', dpi=200, bbox_inches='tight')
plt.close()
```
